```python
import pandas as pd
import os
pd.set_option('display.max_columns', 500)
data_path = '/Users/kylieleblancKylie/domino/dofis/data/'
```

```python
data = pd.read_csv(os.path.join(data_path, 'clean', 'gdid.csv'))
sample = data.sample(1000)
list(sample.columns)
subjects = (list(sample.filter(regex = ("_avescore"))))
variables = ['campus', 'year'] + subjects
sample = sample[variables]
sample.sample(8)
```

```python
reshape = pd.melt(sample, id_vars = ['campus', 'year'])
reshape = reshape.rename(columns = {'variable': 'test', 'value': 'score'})
reshape = reshape.dropna(axis = 0)
reshape.sample(5)
```

```python
means = []
sds = []
for var in subjects:
    mean = reshape[(reshape.test == var) & (reshape.year == 2015)].score.mean()
    means.append(mean)
    sd = reshape[reshape.test == var].score.std()
    sds.append(sd)

means_sds = pd.DataFrame(list(zip(subjects, means, sds)), 
               columns =['test', 'test_mean', 'test_std']) 
means_sds
```

```python
df = reshape.merge(means_sds, left_on = 'test', right_on = 'test')
df.sample(5)
```

```python
df['score_std'] = (df.score - df.test_mean)/df.test_std
df
```
