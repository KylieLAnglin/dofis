### Add column called correct. Put correct year in this column. Save as fix_dates_corrected.csv. Then upload.

```python
#TODO move this to cleaning file
old_year = pd.read_csv(os.path.join(data_path,'doi_dates_and_exemptions.csv'), sep=",")
new_year = pd.read_csv(os.path.join(data_path, 'fix_dates_corrected.csv'), sep=",")
new_year = new_year[['title', 'correct']]
new_year = new_year.rename(columns = {'correct':'doi_year'})
new_year = new_year[['title', 'doi_year']]
new_year.head()
```

```python
data = old_year.merge(new_year, left_on = 'title', right_on = 'title', how = 'left')
data.to_csv(os.path.join(data_path,'doi_dates_and_exemptions.csv'),
                  sep=",")
data.head()
```
