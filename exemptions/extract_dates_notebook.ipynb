```python
import os
import pandas as pd
import extract_dates
import spacy
```

```python
docs = pd.read_csv(os.path.join(os.getcwd(), '..', '..', 'data', 'text_narrowed.csv'))
```

```python
docs
```

```python
output_dir = os.path.join(os.getcwd(), '..', '..', 'data', 'date_term_classifier')
nlp = spacy.load(output_dir)
```

```python
start_dates = []
date_phrases = []
p_terms = []
for text in docs.text:
    date, phrase, p = extract_dates.get_term_date_and_phrase(text, output_dir)
    start_dates.append(date)
    date_phrases.append(phrase)        
    p_terms.append(p)
```

```python
docs['term_date'] = start_dates
docs['term_phrase'] = date_phrases
docs['term_p'] = p_terms
docs.head()
```

```python
# finalize plan date
```

```python
output_dir = os.path.join(os.getcwd(), '..', '..', 'data', 'date_finalize_classifier')
nlp = spacy.load(output_dir)
```

```python
finalize_dates = []
finalize_phrases = []
finalize_p = []
for text in docs.text:
    date, phrase, p = extract_dates.get_finalize_date_and_phrase(text, output_dir)
    finalize_dates.append(date)
    finalize_phrases.append(phrase)
    finalize_p.append(p)
```

```python
docs['finalize_date'] = finalize_dates
docs['finalize_phrase'] = finalize_phrases
docs['finalize_p'] = finalize_p
docs.head(10)
```

```python
docs.loc[docs.term_date >= docs.finalize_date, 'date'] = docs.term_date
docs.loc[docs.term_date < docs.finalize_date, 'date'] = docs.finalize_date
docs.loc[docs.term_date >= docs.finalize_date, 'date_phrase'] = docs.term_phrase
docs.loc[docs.term_date < docs.finalize_date, 'date_phrase'] = docs.finalize_phrase
docs.loc[docs.term_date >= docs.finalize_date, 'date_p'] = docs.term_p
docs.loc[docs.term_date < docs.finalize_date, 'date_p'] = docs.finalize_p
docs.head()
```

```python
docs_dates = docs[['title', 'date', 'date_phrase', 'date_p']]
```

```python
docs_dates.to_csv(os.path.join(os.getcwd(), '..', '..', 'data', 'doi_dates_scraped.csv'))
```
