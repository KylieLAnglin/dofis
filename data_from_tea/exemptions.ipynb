```python
import pandas as pd
import os
from library import start
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
def make_list(string):
    new_list = [i.replace('[','').replace(']','') for i in string.split(', ')]
    return new_list
```

```python
my_df = pd.read_csv('/Users/kylieleblancKylie/domino/dofis/data/plans/doi_exemptions_list.csv')
my_df = my_df[['title', 'link', 'possible_laws']]

my_df['possible_laws'] = my_df['possible_laws'].apply(make_list)
mlb = MultiLabelBinarizer()
my_df = my_df.join(pd.DataFrame(mlb.fit_transform(my_df.pop('possible_laws')),
                          columns=mlb.classes_,
                          index=my_df.index))
pd.options.display.max_columns = 150
my_df.head(10)
```
